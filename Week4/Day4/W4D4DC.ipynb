Download and import the Data Science Job Salary dataset.

```python
import pandas as pd
```

```python
df = pd.read_csv(r'/content/datascience_salaries.csv')
```

```python
df
```

Normalize the â€˜salaryâ€™ column using Min-Max normalization.

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary']])

df
```

Aggregate data by â€˜experience_levelâ€™, calculating average and median salaries.

```python
salary_mean = df.groupby('experience_level',as_index=False).agg({'salary':'mean'})
salary_median = df.groupby('experience_level',as_index=False).agg({'salary':'median'})
salary_by_experience = salary_mean.merge(salary_median, on = 'experience_level')
salary_by_experience.rename(columns={'salary_x':'mean_salary', 'salary_y':'median_salary'}, inplace=True)
salary_by_experience['mean_salary'].round(1)
salary_by_experience
```

Implement dimensionality reduction on the dataset.

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
from sklearn.preprocessing import StandardScaler

features = ['salary', 'salary_normalized']

# Separating out the features
x = df.loc[:, features].values

# Separating out the target
y = df.loc[:,['experience_level']].values

# Standardizing the features
x = StandardScaler().fit_transform(x)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

principalComponents = pca.fit_transform(x)

principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf
```

```python
finalDf = pd.concat([principalDf, df[['experience_level']]], axis = 1)
finalDf
```

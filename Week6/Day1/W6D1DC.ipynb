Use NumPy to generate a synthetic dataset representing average monthly temperatures (in degrees Celsius) for 12 months across 10 different cities. The temperatures should range from -5 to 35 degrees.

Convert this NumPy array into a Pandas DataFrame, adding city names as index and months as columns.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Generate the temperature data
temperature_data = np.random.uniform(-5, 35, size=(10, 12))
```

```python
# Define city names and month names
cities = ['Moscow', 'Paris', 'Tokio', 'New York', 'Toronto', 'Rome', 'Kyiv', 'Minsk', 'Oslo', 'Helsinki']
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
# Create a DataFrame
temperature_df = pd.DataFrame(temperature_data, index=cities, columns=months)

print(temperature_df)
```

2. Data Analysis:



Hint 1: Calculate the annual average temperature using DataFrame.mean(axis).

Hint 2: Find the city with the highest and lowest average temperature using idxmax() and idxmin() methods.

Calculate the annual average temperature for each city.

Identify the city with the highest and lowest average temperature for the year.

```python
# Calculate annual average temperature
annual_avg_temperature = temperature_df.mean(axis=1)

print("Annual Average Temperature: \n")
print(annual_avg_temperature)
```

```python
# Identify the city with the highest and lowest average temperature
city_highest_avg_temp = annual_avg_temperature.idxmax()
print("City with the Highest Average Temperature:", city_highest_avg_temp)

city_lowest_avg_temp = annual_avg_temperature.idxmin()
print("City with the Lowest Average Temperature:", city_lowest_avg_temp)
```

3. Data Visualization:



Hint 1: Use plt.plot() for each city’s temperature trends. Adjust the linewidth and linestyle for the highlighted cities.

Hint 2: Use plt.title(), plt.xlabel(), plt.ylabel(), and plt.legend() to add details to your plot.

Use Matplotlib to create a line plot showing the monthly temperature trends for each city.

Highlight the city with the highest and lowest average annual temperature in a different color or line style.

```python
# Plot temperature trends for each city
plt.figure(figsize=(12, 6))

for city in cities:
    if city == city_highest_avg_temp:
        plt.plot(temperature_df.columns, temperature_df.loc[city], label=city, linestyle='--', linewidth=3)
    elif city == city_lowest_avg_temp:
        plt.plot(temperature_df.columns, temperature_df.loc[city], label=city, linestyle=':', linewidth=3)
    else:
        plt.plot(temperature_df.columns, temperature_df.loc[city], label=city, linewidth=1)

# Add labels and title
plt.title('Monthly Temperature Trends')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.legend()

plt.show()
```


ðŸŒŸ Exercise 1 : Linear Regression Analysis

Objective: Apply linear regression to a dataset and interpret the results.

Task: Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.

```python
house_sizes = [50, 70, 80, 100, 120]
house_prices = [150000, 200000, 210000, 250000, 280000]
```

```python
# What is the slope and intercept of the regression line?

# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes, house_prices)

# The slope and intercept of the regression line
print("Slope:", slope)
print("Intercept:", intercept)
```

```python
# Use linear regression
sns.regplot(x = house_sizes, y = house_prices)
```

```python
# Predict the price of a house that is 90 square meters.
# y = intercept + slope*x

house_90_meters_price = intercept + slope * 90
round(house_90_meters_price,0)
```



---

In the context of housing prices, the slope in the linear regression equation represents the rate of change in house prices for each unit increase in house size. Specifically, it indicates how much the price of a house is expected to increase (or decrease) for every additional square meter of space.

ðŸŒŸ Exercise 2 : Understanding ANOVA

Objective: Test understanding of ANOVA and its application.

Task: Three different fertilizers are applied to three separate groups of plants to test their effectiveness. The growth in centimeters is recorded.

```python
fertilizer_1 = [5, 6, 7, 6, 5]
fertilizer_2 = [7, 8, 7, 9, 8]
fertilizer_3 = [4, 5, 4, 3, 4]
```

```python
# Perform an ANOVA test on the given data. What are the F-value and P-value?
f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)
```

**Based on the P-value, do the fertilizers have significantly different effects on plant growth?**

The p-value is very small, which is much smaller than the commonly used significance level of 0.05. When the p-value is less than the significance level, it is an indication that we can reject the null hypothesis.

In the context of plant growth and fertilizer effects, this suggests that there are significant differences among the groups treated with different fertilizers. In other words, the fertilizers appear to have significantly different effects on plant growth.

**Explain what would happen if the P-value were greater than 0.05.**

In the context of plant growth and fertilizer effects, if the P-value is greater than 0.05, you might conclude that there is no statistically significant difference in the effects of different fertilizers on plant growth. This could lead to the interpretation that, based on the available evidence, the fertilizers have similar effects.

Example: Calculating the probability of getting exactly 5 heads in 10 coin flips.

ðŸŒŸ Exercise 4 : Correlation Coefficients

Calculate the Pearson and Spearman correlation coefficients between two variables in a dataset.

```python
data = pd.DataFrame({'age': [23, 25, 30, 35, 40], 'income': [35000, 40000, 50000, 60000, 70000]})
data
```

```python
# Pearson Correlation Coefficient
st.pearsonr(data['age'], data['income'])
```

```python
# Spearman Correlation Coefficient
st.spearmanr(data['age'], data['income'])
```

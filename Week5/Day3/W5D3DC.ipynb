#Daily Challenge: Mastering Data Visualization


Dataset: Use the ‘Iris’ dataset available in seaborn (use seaborn.load_dataset('iris')).


```python
import seaborn as sns
import matplotlib.pyplot as plt
from plotnine import *
import plotly.express as px
```

```python
# Load the dataset
iris = sns.load_dataset('iris')
```


Plotnine Task: Create a scatter plot of sepal length vs. sepal width, color-coded by species.


```python
# Create a scatter plot using plotnine
scatter_plot = ggplot(iris, aes(x='sepal_length', y='sepal_width', color='species'))+geom_point()+theme_minimal()+facet_wrap('~species')

scatter_plot.draw()
```


Matplotlib Task: Generate a histogram for petal lengths.


```python
# Create a histogram for petal lengths
plt.figure(figsize=(8, 6))
plt.hist(iris['petal_length'], bins=20, color='skyblue', edgecolor='black')

plt.title('Histogram of Petal Lengths')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')

plt.show()
```


Seaborn Task: Produce a pair plot to show pairwise relationships in the dataset.

```python
# Create a pair plot
sns.pairplot(iris, hue='species', palette='viridis', markers=["o", "s", "D"])

plt.show()
```


Plotly Task: Create an interactive 3D scatter plot mapping sepal length, sepal width, and petal length, color-coded by species.

```python
# Create a scatter plot
sns.set(style="whitegrid")
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', size='petal_length', sizes=(15, 200), data=iris)

plt.title('Scatter Plot of Sepal Length vs Sepal Width, Color-coded by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

plt.show()
```

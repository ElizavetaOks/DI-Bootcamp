Data visualization is crucial in data analysis because it allows complex datasets to be presented in a visual format, making patterns, trends, and insights more easily understandable. # üåü Exercise 2 : Creating A Line Plot For Temperature Variation

Create a simple line plot using Matplotlib that represents the temperature variations over a week.

Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.

```python
import matplotlib.pyplot as plt

Day = [i for i in range(1, 8)]
Temperature_F = [67, 71, 75, 71, 73, 74, 80]

plt.plot(Day, Temperature_F, color='green', linestyle='-.', linewidth=2, marker='o',
         markerfacecolor='blue', markersize=6)
plt.xlabel('Day')
plt.ylabel('Temperature (¬∞F)')
plt.title('The Temperature variations over a Week')

plt.show()
```

# üåü Exercise 3 : Visualizing Monthly Sales With A Bar Chart

Generate a bar chart using Matplotlib to visualize the monthly sales data for a retail store.

Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]) and label the x-axis as ‚ÄúMonth‚Äù and the y-axis as ‚ÄúSales Amount ($)‚Äù.

```python
Month = [i for i in range(1, 13)]
Sales_Amount = [6700, 7100, 7050, 7001, 7300, 5874, 8000, 5014, 9214, 4583, 7000, 6850]

plt.plot(Month, Sales_Amount, color='blue', linestyle='dotted', linewidth=2, marker='h',
         markerfacecolor='blue', markersize=8)
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('The Monthly Sales')

plt.show()
```

```python
import pandas as pd
```

```python
df.drop_duplicates()
```


Identify the category that had the highest revenue and calculate how much revenue it generated.



```python
df.groupby('category', as_index = False).agg({'revenue':'sum'}).sort_values(by = 'revenue', ascending=False)
```

Determine the average revenue per sale in the dataset.


```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
df['quarter'] = df.date.dt.quarter
```

```python
quarter_mapping = {1: 'Q1', 2: 'Q2', 3: 'Q3', 4: 'Q4'}
df['quarter'] = df['quarter'].map(quarter_mapping)
```

```python
df
```

```python
df.groupby('quarter', as_index = False).agg({'revenue':'mean'})
```

Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

```python
quarter_revenue = df.groupby('quarter', as_index = False).agg({'revenue':'sum'})
quarter_revenue
```

```python
import matplotlib.pyplot as plt

x = [i for i in quarter_revenue['quarter']]
y = [i for i in quarter_revenue['revenue']]

plt.bar(x,y, color=['red', 'yellow', 'orange', 'blue'], edgecolor='black', hatch='.')
plt.ylabel('Revenue in $')
plt.title('Total revenue per quarter')

plt.show()
```


# üåü Exercise 5 : Data Visualisation Using MatPlotLib

Create a simple line plot using Matplotlib. Plot the function y = x^2 for x values ranging from -10 to 10. Add a title and label the x and y axes.


```python
x = [i for i in range(-10, 11)]
y_2 = [i**2 for i in x]
plt.plot(x, y_2, label='x squared 2', color='pink',linestyle='-', linewidth=2, marker='o',
         markerfacecolor='yellow', markersize=6)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Simple Line Plot')
plt.legend()
plt.show()
```


Generate a bar chart to display the sales data of four different products (A, B, C, D) in a week. The sales values are 15, 30, 45, and 20 respectively. Label the chart appropriately.


```python
products = ['A', 'B', 'C', 'D']
sales = [15, 30, 45, 20]

plt.bar(products, sales, color=['lightblue'], edgecolor='black', hatch='*')
plt.title('Products Sales Data')
plt.xlabel('Products')
plt.ylabel('Sales values')
plt.show()
```


Create a pie chart representing the following data about favorite fruits of a group of people: Apples (40%), Bananas (30%), Cherries (20%), and Dates (10%). Customize the chart with different colors for each fruit and add a legend.

```python
fruits = ['Apples', 'Bananas', 'Cherries', 'Dates']
people = [40,30,20,10]

plt.figure(figsize=(6, 6))
explode = (0.05, 0.05, 0.05, 0.05)
plt.pie(people, labels=fruits, autopct='%1.1f%%', explode=explode)
plt.title('Favorite fruits of a group of people')
plt.show()
```

# üåü Exercise 2 : Creating A Line Plot For Temperature Variation

Create a simple line plot using Matplotlib that represents the temperature variations over a week.

Use a list of temperature values for each day of the week (e.g., [72, 74, 76, 80, 82, 78, 75]) and label the x-axis as ‚ÄúDay‚Äù and the y-axis as ‚ÄúTemperature (¬∞F)‚Äù.

```python
import matplotlib.pyplot as plt

Day = [i for i in range(1, 8)]
Temperature_F = [67, 71, 75, 71, 73, 74, 80]

plt.plot(Day, Temperature_F, color='green', linestyle='-.', linewidth=2, marker='o',
         markerfacecolor='blue', markersize=6)
plt.xlabel('Day')
plt.ylabel('Temperature (¬∞F)')
plt.title('The Temperature variations over a Week')

plt.show()
```

# üåü Exercise 3 : Visualizing Monthly Sales With A Bar Chart

Generate a bar chart using Matplotlib to visualize the monthly sales data for a retail store.

Create a list of sales values for each month (e.g., [5000, 5500, 6200, 7000, 7500]) and label the x-axis as ‚ÄúMonth‚Äù and the y-axis as ‚ÄúSales Amount ($)‚Äù.

```python
Month = [i for i in range(1, 13)]
Sales_Amount = [6700, 7100, 7050, 7001, 7300, 5874, 8000, 5014, 9214, 4583, 7000, 6850]

plt.plot(Month, Sales_Amount, color='blue', linestyle='dotted', linewidth=2, marker='h',
         markerfacecolor='blue', markersize=8)
plt.xlabel('Month')
plt.ylabel('Sales Amount ($)')
plt.title('The Monthly Sales')

plt.show()
```

```python
df.groupby('category', as_index = False).agg({'revenue':'sum'}).sort_values(by = 'revenue', ascending=False)
```

Determine the average revenue per sale in the dataset.


```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'])
df['quarter'] = df.date.dt.quarter
```

```python
quarter_mapping = {1: 'Q1', 2: 'Q2', 3: 'Q3', 4: 'Q4'}
df['quarter'] = df['quarter'].map(quarter_mapping)
```

```python
df
```

```python
df.groupby('quarter', as_index = False).agg({'revenue':'mean'})
```

Create a bar chart to visualize the total revenue generated in each quarter of the year (Q1, Q2, Q3, Q4).

```python
quarter_revenue = df.groupby('quarter', as_index = False).agg({'revenue':'sum'})
quarter_revenue
```

```python
import matplotlib.pyplot as plt

x = [i for i in quarter_revenue['quarter']]
y = [i for i in quarter_revenue['revenue']]

plt.bar(x,y, color=['red', 'yellow', 'orange', 'blue'], edgecolor='black', hatch='.')
plt.ylabel('Revenue in $')
plt.title('Total revenue per quarter')

plt.show()
```

# Exercise 2: Customer Lifetime Value (CLV) Analysis

```python
# Create a new dataframe summarizing purchase frequency, average purchase value, and total sales per customer.
new_df = df.groupby('Customer ID', as_index = False).agg(purchase_frequency=('Order ID','nunique'),
                                                         avarage_purchase =('Sales', 'mean'),
                                                         total_sales = ('Sales', 'sum'),
                                                         first_date = ('Order Date','min'),
                                                         last_date = ('Order Date', 'max'))
new_df.head()
```

```python
# Calculate CLV for each customer
new_df.dtypes
```

```python
new_df['lifespan'] = new_df['last_date']-new_df['first_date']
```

```python
new_df['clv'] = new_df['purchase_frequency']*new_df['avarage_purchase']*(new_df['lifespan'].dt.days/365)
new_df['clv'] = round(new_df['clv'], 2)
```

```python
#Identify top customers by CLV and visualize their contribution.
sorted_df = new_df.sort_values(by = 'clv', ascending=False).head(10)
plt.bar(sorted_df['Customer ID'], sorted_df['clv'])
plt.xlabel('Customers')
plt.xticks(rotation=90)
plt.ylabel('CLV')
plt.title('Top 10 Customers by CLV')
plt.legend()
plt.show()
```

```python
import pandas as pd

# Assuming your DataFrame is named 'df'
# If not, replace 'df' with the actual name of your DataFrame

# Convert 'Order Date' to datetime
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract 'Order_Year' from 'Order Date'
df['Order_Year'] = df['Order Date'].dt.year

# Calculate purchase frequency, average purchase value, and total sales per customer
customer_summary = df.groupby('Customer ID').agg({
    'Order ID': 'nunique',            # Purchase frequency
    'Sales': ['mean', 'sum'],         # Average purchase value and total sales
    'Order_Year': 'max'               # Latest purchase year
}).reset_index()

# Rename columns for clarity
customer_summary.columns = ['Customer ID', 'Purchase Frequency', 'Average Purchase Value', 'Total Sales', 'Latest Purchase Year']

# Calculate CLV for each customer (assuming a simple CLV calculation)
# For simplicity, CLV = Average Purchase Value * Purchase Frequency
customer_summary['CLV'] = customer_summary['Average Purchase Value'] * customer_summary['Purchase Frequency']

# Identify top customers by CLV
top_customers = customer_summary.sort_values(by='CLV', ascending=False).head(10)

# Visualize the contribution of top customers
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(top_customers['Customer ID'], top_customers['CLV'], color='blue')
plt.xlabel('Customer ID')
plt.ylabel('Customer Lifetime Value (CLV)')
plt.title('Top Customers by CLV')
plt.xticks(rotation=45)
plt.show()
```

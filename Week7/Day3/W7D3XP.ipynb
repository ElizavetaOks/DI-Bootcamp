##ðŸŒŸ Exercise 1: Implementing And Comparing Normalization Techniques

Create a synthetic dataset with two features: FeatureA and FeatureB, where FeatureA is in the range of [0, 1000] and FeatureB is in the range of [0, 10].
Apply Min-Max Scaling to this dataset and visualize the results using a scatter plot.
Apply Z-score Normalization to the same dataset and visualize these results as well.
Write a short analysis comparing the effects of both normalization techniques on the dataset. Discuss which technique might be more appropriate based on the distribution and range of the data.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler, StandardScaler
import pandas as pd
import seaborn as sns
```

```python
# Generate synthetic dataset
np.random.seed(42)
num_samples = 1000
FeatureA = np.random.rand(num_samples) * 1000
FeatureB = np.random.rand(num_samples) * 10
```

```python
# Apply Min-Max Scaling
min_max_scaler = MinMaxScaler()
min_max_scaled_data = min_max_scaler.fit_transform(np.column_stack((FeatureA, FeatureB)))
FeatureA_minmax, FeatureB_minmax = min_max_scaled_data[:, 0], min_max_scaled_data[:, 1]

# Create a scatter plot for Min-Max Scaled dataset
plt.subplot(1, 3, 2)
plt.scatter(FeatureA_minmax, FeatureB_minmax, alpha=0.5)
plt.title('Min-Max Scaling')
plt.xlabel('FeatureA (Min-Max Scaled)')
plt.ylabel('FeatureB (Min-Max Scaled)')

plt.show()
```

```python
# Apply Z-score Normalization
z_score_scaler = StandardScaler()
z_score_normalized_data = z_score_scaler.fit_transform(np.column_stack((FeatureA, FeatureB)))
FeatureA_zscore, FeatureB_zscore = z_score_normalized_data[:, 0], z_score_normalized_data[:, 1]

# Create a scatter plot for Z-score Normalized dataset
plt.subplot(1, 3, 3)
plt.scatter(FeatureA_zscore, FeatureB_zscore, alpha=0.5)
plt.title('Z-score Normalization')
plt.xlabel('FeatureA (Z-score Normalized)')
plt.ylabel('FeatureB (Z-score Normalized)')

plt.tight_layout()
plt.show()
```

* If the dataset has outliers and you want to preserve their influence, Min-Max Scaling might be more appropriate.
* If the dataset is approximately normally distributed and doesn't have significant outliers, Z-score Normalization may be suitable.

##ðŸŒŸ Exercise 4 : Identifying Outliers With Box Plots And Z-Scores

Load the Boston Housing Dataset into a Python environment.
Use box plots to visually identify potential outliers in various features like â€˜CRIMâ€™ (crime rate), â€˜RMâ€™ (average number of rooms per dwelling), and â€˜MEDVâ€™ (Median value of owner-occupied homes in $1000â€™s).
Calculate Z-scores for these features and identify data points that have a Z-score higher than 3 or lower than -3.
Reflect on how outliers identified by box plots and Z-scores overlap or differ.


```python
# Load the Boston Housing Dataset
df=pd.read_csv(r'/content/HousingData.csv')
df.head()
```

```python
df.dtypes
```

```python
# Extract the features of interest
features_of_interest = ['CRIM', 'RM', 'MEDV']
subset_df = df[features_of_interest]
```

```python
import pylab

# Visualize potential outliers using box plots
plt.figure(figsize=(12, 5))
plt.subplot (2, 2, 1)
sns.boxplot(x=subset_df["CRIM"])
pylab.title ("Box plot of CRIM")

plt.subplot (2, 2, 3)
sns.boxplot(x=subset_df["RM"])
pylab.title ("Box plot of RM")

plt.subplot (1, 2, 2)
sns.boxplot(x=subset_df["MEDV"])
pylab.title ("Box plot of MEDV")

plt.tight_layout()
plt.show()
```

```python
# Calculate Z-scores for selected features
z_score_scaler = StandardScaler()
z_scores = z_score_scaler.fit_transform(subset_df)
```

```python
# Identify data points with Z-scores higher than 3 or lower than -3
outliers_mask = np.abs(z_scores) > 3
outliers_indices = np.argwhere(outliers_mask)
```

```python
# Display indices of outliers
print("Indices of outliers:")
print(outliers_indices)
```

```python
# Display values of outliers
print("\nValues of outliers:")
print(subset_df.values[outliers_indices[:, 0], :])
```

Reflecting on the results, we can compare the outliers identified by box plots and Z-scores. Box plots provide a visual representation of the spread of the data and highlight potential outliers based on interquartile range. Z-scores, on the other hand, provide a numerical measure of how far each data point is from the mean in terms of standard deviations.

## ðŸŒŸ Exercise 5 : Handling Outliers In A Large Dataset

Load the New York City Airbnb dataset.
Identify outliers in the â€˜priceâ€™ column using the IQR method.
Apply three different outlier handling techniques: trimming, capping, and transformation.
Analyze the effect of each technique on the distribution of the â€˜priceâ€™ feature.
Discuss the pros and cons of each technique in the context of this dataset.


```python
df_NY = pd.read_csv(r'/content/AB_NYC_2019.csv')
df_NY.head()
```

```python
# Identify outliers in the â€˜priceâ€™ column
Q1 = df_NY['price'].quantile(0.25)
Q3 = df_NY['price'].quantile(0.75)
IQR = Q3 - Q1

# Defining bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtering the outliers
outliers = df_NY[(df_NY['price'] < lower_bound) | (df_NY['price'] > upper_bound)]
```

```python
# Apply outlier handling techniques
# 1. Trimming: Remove outliers
trimmed_data = df_NY[(df_NY['price'] >= lower_bound) & (df_NY['price'] <= upper_bound)]

# 2. Capping: Cap outliers at the upper bound
capped_data = df_NY.copy()
capped_data['price'] = capped_data['price'].clip(lower=None, upper=upper_bound)

# 3. Transformation: Log-transform the 'price' feature
transformed_data = df_NY.copy()
transformed_data['price'] = np.log1p(transformed_data['price'])
```

```python
# Visualize the effects of each technique
plt.figure(figsize=(15, 6))

# Original distribution
plt.subplot(1, 4, 1)
sns.histplot(df_NY['price'], bins=50, kde=True)
plt.title('Original Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Trimming
plt.subplot(1, 4, 2)
sns.histplot(trimmed_data['price'], bins=50, kde=True)
plt.title('Trimmed Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Capping
plt.subplot(1, 4, 3)
sns.histplot(capped_data['price'], bins=50, kde=True)
plt.title('Capped Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')

# Transformation
plt.subplot(1, 4, 4)
sns.histplot(transformed_data['price'], bins=50, kde=True)
plt.title('Log-Transformed Distribution')
plt.xlabel('Log')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
